```python
import gradio as gr
import torch
from transformers import pipeline
import ast
import re
import black
import autopep8
```

```python
model = pipeline(
    "text-generation",
    model="ibm-granite/granite-3.3-2b-instruct",
    device=0 if torch.cuda.is_available() else -1
)
```

```python
def detect_python_code(text):
    patterns = [
        r'def\s+\w+\(',
        r'class\s+\w+',
        r'import\s+\w+',
        r'from\s+\w+\s+import',
        r'if\s+.*:',
        r'for\s+.*\s+in\s+',
        r'while\s+.*:'
    ]

    for p in patterns:
        if re.search(p, text):
            return True
    return False
```

```python
def check_syntax(code):
    try:
        ast.parse(code)
        return "No syntax error"
    except SyntaxError as e:
        return f"Syntax Error at line {e.lineno}: {e.msg}"
```

```python
def chatbot(user_input):

    if detect_python_code(user_input):
        result = check_syntax(user_input)
        return "Python code detected\n" + result

    response = model(user_input, max_new_tokens=100)

    return response[0]["generated_text"]


demo = gr.Interface(
    fn=chatbot,
    inputs="textbox",
    outputs="textbox",
    title="Syntax Surgeon Chatbot"
)

demo.launch()
```
